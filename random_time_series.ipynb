```python
time_series_return_list = numpy.random.normal(0.024, 0.755, 1000)
value = 10500
time_series = []
high_lag = []
low_lag = []
lag = 100

for day_return in time_series_return_list:
    value = (value + value * (day_return / 100))
    time_series.append(value)
    
    if len(time_series) > lag:
        high_lag.append(numpy.max(time_series[-lag:]))
        
    if len(time_series) > lag:
        low_lag.append(numpy.min(time_series[-lag:]))

high_lag = numpy.full(lag, high_lag[0]).tolist() + high_lag
low_lag = numpy.full(lag, low_lag[0]).tolist() + low_lag
pyplot.plot(time_series, label='series')
pyplot.plot(low_lag, label='low')
output = pyplot.plot(high_lag, label='high')
```
